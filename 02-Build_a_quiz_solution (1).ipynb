```python
def ask_question(nb_attempts_left, question_sentence, right_answer): # three arguments
    if nb_attempts_left > 0:
        answer = input(question_sentence)
        answer = answer.lower()
        while answer != right_answer:
            nb_attempts_left -=1
            print("Sorry, you have {} chances left".format(nb_attempts_left))
            if nb_attempts_left == 0:
                print("Oh no, you lost the quiz...")
                break
            answer = input(question_sentence)
            answer = answer.lower()
    print('')
    return nb_attempts_left

questions_answers_list = [
    ("How many times France soccer team won the World Cup?", "2"),
    ("When was Apple founded?", "1976"),
    ("Who founded SpaceX?", "elon musk")
]

nb_attempts = 3

for question, answer in questions_answers_list :
    nb_attempts = ask_question(nb_attempts, question, answer) # calling the function

if nb_attempts > 0:
    print("Well done, you have won!")
```

```python
#FLORIAN G.

#We define a function that will proceed each quesiton, a question is a list [question,answer]
def party(lives,phase):
    question = phase[0]
    answer = phase[1]
    while lives > 0 :
        myanswer = input(question)
        myanswer = myanswer.lower()
        if myanswer == answer.lower():
            print ("Your answer is : {}. Good job! This is the right answer".format(myanswer))
            return True
        else: 
            lives -=1
            print ("Your answer is : {}. Too bad! that is not the correct answer. You have {} lives left. Try again".format(myanswer, lives))
        if lives == 0:
            print("Oh no! You have lost the game...")
            return False
#We create a class Game that will create one instance per game. The liste in parameter is a list of lists, each element will be a list [question,answer]
class Game ():
    def __init__(self, lives, liste):
        self.lives = lives
        self.liste = liste
    def play(self): #The method that will lauch the play 
        for phase in self.liste:
            result = party(self.lives, phase)
            if result == False:
                break
        print ("You have won the game with {} lives left".format(self.lives))
        
phase1 = ["question1","answer1"]
phase2 = ["question2","answer2"]
game1 = Game(3,[phase1,phase2])
game1.play()
```

```python
#SEBASTIAN P.

class Quiz():
    
    def __init__(self, quiz_list):
        self.quiz_list = quiz_list
        self.correct_answers = 0
        self.wrong_answers = 3

    def ask_question(self, question, answer):
        user_answer = input(f"{question}").lower()
        if user_answer == answer:
            print("Good job! This is the right answer")
            return True
        else:
            print("Too bad! that is not the correct answer")
            return False
    
    def show_results(self):
        print(f"You answered {self.correct_answers} out of {len(self.quiz_list)} questions correctly.")
        if self.correct_answers == len(self.quiz_list):
            print("Well done! You are the quiz winner!")
    
    def play_quiz(self):
        print("Voici notre quiz, tu as trois chances !")
        for i in self.quiz_list:
            check_answer = False
            while (check_answer is False) and (self.wrong_answers > 0):
                check_answer = self.ask_question(i['question'], i['answer'])
                if check_answer:
                    self.correct_answers += 1
                    break
                elif self.wrong_answers == 0:
                    break
                else:
                    self.wrong_answers -= 1
                    print(f"Dommage! Il te reste {self.wrong_answers} chance(s).")
        self.show_results()


quiz = [
    {'question': 'Combien de fois la France a gagné la coupe du monde ? ', 'answer': '2'},
    {'question': 'Quand a été fondé Apple ? ', 'answer': '1976'},
    {'question': 'Qui a fondé SpaceX ? ', 'answer': 'elon musk'},
]

quiz = Quiz(quiz)

quiz.play_quiz()
```
